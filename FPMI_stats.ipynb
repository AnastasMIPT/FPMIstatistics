```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./FPMIstatistics.csv')
```

```python
study_fields = ['ПМИ (обычный)', 'КТ', 'ИВТ', 'ПМ']
```

## Работа и стажировки

```python
plt.figure(figsize=(15, 10))
sns.countplot(x='Закончил курс', hue='Стажируешься(лся) / работаешь(л)?', data=df)
plt.title('Количество стажирующихся и не стажирующихся людей по курсам', size = 15)
plt.show()
```

#### Мои наблюдения

Со второго курса большинство ребят начинает стажироваться или работать.

```python
plt.figure(figsize=(15, 10))
sns.countplot(x='Направление', hue='Стажируешься(лся) / работаешь(л)?', data=df[df['Закончил курс'] == 2])
plt.title('Количество стажирующихся и не стажирующихся людей с второго курса по направлениям', size = 15)
plt.show()
```

#### Мои наблюдения

На ПМ на втором курсе стажируется все еще меньшая часть людей.

#### То же самое, только отнормированное

```python
plt.figure(figsize=(16,12))
filters = []
course = 2
for field in study_fields:
    filters.append((df['Направление'] == field) & (df['Закончил курс'] == course))
for i in range(4):
    plt.subplot(1, 4, i + 1)
    label = list(df[filters[i]]['Стажируешься(лся) / работаешь(л)?'].value_counts().index)
    plt.pie(df[filters[i]]['Стажируешься(лся) / работаешь(л)?'].value_counts(), labels=label)
    plt.title(study_fields[i])
plt.show()
```

#### Мои наблюдения

Наибольшая доля стажирующихся ребят-второкурсников - на ИВТ. Наименьшая - на ПМ.

```python
plt.figure(figsize=(14, 7))
plot = sns.countplot(data = df[df['Закончил курс'] == 2], x='Если да, то где?')
plot.set_xticklabels(plot.get_xticklabels(), rotation=50)
plt.title('Компании, в которых стажируются второкурсники ФПМИ', size=15)
plt.show()
```
